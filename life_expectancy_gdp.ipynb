```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
# Load the data into life_data
life_data = pd.read_csv('all_data.csv')
print(life_data.head(5))
```

```python
unique_countries = life_data['Country'].unique()
print(unique_countries)
```

```python
for country in unique_countries:
    sns.scatterplot(x = 'Life expectancy at birth (years)', y = 'GDP', hue = 'Year', data = life_data[life_data['Country'] == country])
    plt.title(f'Life Expectancy in {country} Over Years')
    plt.xlabel('Life expectancy at birth (years)')
    plt.ylabel('GDP')
    plt.show()
    plt.clf()
```

```python
life_data_age_means = life_data.drop("Year", axis = 1).groupby("Country").mean().reset_index()
sns.barplot(x = 'Life expectancy at birth (years)', y = 'Country', data = life_data_age_means)
plt.title('Mean life expectancy of country from 2000 - 2015')
plt.show()
plt.clf()
```

```python
life_data_age_means = life_data.drop("Year", axis = 1).groupby("Country").mean().reset_index()
sns.barplot(x = 'GDP', y = 'Country', data = life_data_age_means)
plt.title('Mean GDP country from 2000 - 2015')
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=life_data.Year, y=life_data['Life expectancy at birth (years)'], hue=life_data.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.show()
plt.clf()
```
